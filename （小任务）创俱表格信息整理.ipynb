```python
name=[]
xuehao=[]
for i in range(len(a)):
    name_i=[]
    xuehao_i=[]
    name_i.append(df_1['负责人姓名'][i])
    xuehao_i.append(df_1['负责人学号'][i])
    for j in a[i]:
        if len(j)>2:
            #print(j)
            xuehao_i.append(j[-7:])
            if '/' in j:
                name_i.append(j[:-8])
            else:
                name_i.append(j[:-7])
    name.append(name_i)
    xuehao.append(xuehao_i)
```
